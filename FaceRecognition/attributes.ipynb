```python
import pandas as pd
```

```python
### loading dataframe and separating values with commas
df = pd.read_csv("./data/list_attr_celeba.txt", skiprows=2, header=None)
df = pd.DataFrame(df[0].str.split(" "))
df
```

```python
### remove double spaces in attributes list
for img in range(len(df)):
    for attr in df.iloc[img, 0]:
        if attr in ["", " "]:
            df.iloc[img, 0].remove(attr)
```

```python
### generating new corrected dataframe
values = [df.iloc[i, 0][1:] for i in range(len(df))]
index = [df.iloc[j, 0][0] for j in range(len(df))]
attributes = pd.DataFrame(values, index=index)
```

```python
attributes = attributes.replace(to_replace="-1", value=0).replace(to_replace="1", value=1)
attributes
```

```python
### saving corrected file
attributes.to_csv("./data/attr_formated.txt")
```
