```python
# Needed libraries
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Layer
from tensorflow.math import add
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Definition of the layer
class rgb_to_gs(Layer):
    def __init__(self):
        super().__init__() # Initialization inherited from Layer class
    
    # Definition of what the layer performs on the inputs
    def call(self, inputs):
        
        # Converting the DirectoryIterator from ImageDataGenerator.flow_from_directory() to tensor
        temporal = np.array([inputs[0][0]])
        for i in range(1, len(inputs)):
            image = np.array([inputs[i][0]])
            temporal = np.concatenate((temporal, image))
            x = tf.convert_to_tensor(temporal)
            
        return tf.math.add(x[:, :, :, :, :1]*0.21, x[:, :, :, :, 1:-1]*0.72, x[:, :, :, :, -1:]*0.07)
```

```python
path = "./images"
ih = 150  # image height
iw = 150  # image width

# Retrieving the images to process
datagen = ImageDataGenerator(rescale=1./255.)

# batch size 1 to process images individually
images = datagen.flow_from_directory(path, batch_size = 1, target_size=(ih, iw))
```

```python
# verifying that the images are retrieved correctly
fig, axs = plt.subplots(1, 3, figsize=(11, 3))
for i in range(len(images)):
    axs[i].imshow(images[i][0][0])
plt.show()
```

```python
# Using the layer alone
layer = rgb_to_gs()
y = layer(images)

# Checking that the images were processed correctly
fig, axs = plt.subplots(1, 3, figsize=(11, 3))
for i in range(len(y)):
    axs[i].imshow(y[i][0])
plt.show()

# print(y)
# print(y.shape)
```

```python
### similar code but using tf.image.rgb_to_grayscale() function
class rgb2gs(Layer):
    def __init__(self):
        super().__init__()
        
    def call(self, inputs):
        temporal = np.array([inputs[0][0]])
        for i in range(1, len(inputs)):
            image = np.array([inputs[i][0]])
            temporal = np.concatenate((temporal, image))
            x = tf.convert_to_tensor(temporal)
        return tf.image.rgb_to_grayscale(x)
    
layer = rgb2gs()
z = layer(images)

fig, axs = plt.subplots(1, 3, figsize=(11, 3))
for i in range(len(z)):
    axs[i].imshow(z[i][0])  # setting cmap="binary" makes images black and white
plt.show()
```
