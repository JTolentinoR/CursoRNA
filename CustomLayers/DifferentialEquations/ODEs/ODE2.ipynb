```python
import numpy as np
from matplotlib import pyplot as plt

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout,Activation
from tensorflow.keras.optimizers import SGD
```

```python
class ODEsolver(Sequential):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        self.loss_tracker = keras.metrics.Mean(name="loss")
        self.mse = tf.keras.losses.MeanSquaredError()

    @property
    def metrics(self):
          return [self.loss_tracker]

    def train_step(self, data):
        batch_size = tf.shape(data)[0]
        x = tf.random.uniform((batch_size,1), minval=-5, maxval=5)
        x_0 = tf.zeros((batch_size, 1))

        with tf.GradientTape() as tape:
            
            # Here the net requires the second derivative
            with tf.GradientTape() as tape3:
                tape3.watch(x)
                
                # Here the net only requires first derivative
                with tf.GradientTape(persistent=True) as tape2:
                    tape2.watch(x)
                    tape2.watch(x_0)
                    y_pred = self(x, training=True)
                    y_0 = self(x_0, training=True)
                dy = tape2.gradient(y_pred, x)
                dy_0 = tape2.gradient(y_0, x_0)
                
            dyy = tape3.gradient(dy, x)
            ic2 = dy_0 + 0.5

            eq = dyy + y_pred
            ic = y_0 - 1.0

            # this includes both initial conditions
            loss = keras.losses.mean_squared_error(0., eq) + keras.losses.mean_squared_error(0., ic) + keras.losses.mean_squared_error(0., ic2)

        
        grads = tape.gradient(loss, self.trainable_variables)
        self.optimizer.apply_gradients(zip(grads, self.trainable_variables))
        
        self.loss_tracker.update_state(loss)
        
        return {"loss": self.loss_tracker.result()}
```

```python
# model definition
model = ODEsolver()

model.add(Dense(10, activation = 'tanh', input_shape=(1,)))
model.add(Dense(10, activation='tanh'))
model.add(Dense(10, activation='tanh'))
model.add(Dense(1, activation='linear'))

model.summary()
```

```python
x_train = tf.linspace(-5, 5, 200)

model.compile(optimizer='rmsprop', metrics=['loss'])
history = model.fit(x_train, epochs=1500, verbose=0)
model.save("eq2.hdf5")
```

```python
x_test = tf.linspace(-5, 5, 200)
prediction = model.predict(x_test)
```

```python
plt.title('Solution to y\'\'=-y with y(0)=1 and y\'(0)=-0.5')
plt.plot(x_test, prediction, color='black', label='Numerical')
plt.plot(x_test, -0.5*np.sin(x_train)+np.cos(x_train), color='red',  label='Analytical')
plt.legend()
plt.show()
```
